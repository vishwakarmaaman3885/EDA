```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\HP\\Downloads\\House Price India.csv")
```

```python
df.info()
```

```python
df.describe()
```

```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# Q1. Average, Median, and Standard Deviation of House Prices

avg_price = df['Price'].mean()
median_price = df['Price'].median()
std_price = df['Price'].std()

print("Average Price:", avg_price)
print("Median Price:", median_price)
print("Standard Deviation:", std_price)
```

```python
# Q2. Most Common Number of Bedrooms

df['number of bedrooms'].mode()
```

```python
# Q3. Check Whether Price Data is Skewed

df['Price'].skew()
```

```python
# Q4. Average Price for Each Bedroom Count

df.groupby('number of bedrooms')['Price'].mean()
```

```python
# Q5. Relationship Between Area and Price

# (Area without basement)

df[['Area of the house(excluding basement)', 'Price']].corr()
```

```python
# Q6. High Price but Low Area Houses

df[(df['Price'] > df['Price'].quantile(0.90)) &
   (df['Area of the house(excluding basement)'] <
    df['Area of the house(excluding basement)'].quantile(0.25))]
```

```python
# Q7. Average Price Based on Floors and Waterfront

df.groupby(['number of floors', 'waterfront present'])['Price'].mean()
```

```python
# Q8. Minimum and Maximum House Price

print("Minimum Price:", df['Price'].min())

print("Maximum Price:", df['Price'].max())
```

```python
# Q9. Postal Code with Highest Average Price

df.groupby('Postal Code')['Price'].mean().sort_values(ascending=False).head(1)
```

```python
# Q10. Write at least 5 insights from your analysis

# Insight 1
# House prices are positively skewed because a few luxury houses increase the average price.

# Insight 2
# Most houses have a common bedroom count such as 3 or 4 bedrooms.

# Insight 3
# Larger living areas usually increase house prices.

# Insight 4
# Waterfront houses are generally more expensive.

# Insight 5
# Some postal codes have much higher average prices, indicating premium locations.
```
